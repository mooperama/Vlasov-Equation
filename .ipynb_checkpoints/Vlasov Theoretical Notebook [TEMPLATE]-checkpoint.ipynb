## 0. The Plasma Medium and the Ordering Behind the Vlasov Model


The Vlasov equation describes a weakly coupled plasma in the mean-field limit. Its domain is narrower than the phrase "ionized gas" suggests. A gas may contain free charges and still fail to behave as a plasma if collisions with neutrals or binary encounters between particles dominate the motion. The relevant distinction is collective response: a displacement of charge modifies the electric field, and that field acts back on a large number of particles before isolated collisions can erase the disturbance.

The electrostatic part of this collective response is fixed by Gauss's law. For an electron-ion plasma with singly charged ions, the charge density is

$$
\rho(\mathbf r,t)=e n_i(\mathbf r,t)-e n_e(\mathbf r,t), \tag{0.1}
$$

where $e>0$ is the magnitude of the electron charge, $n_i$ is the ion number density, and $n_e$ is the electron number density. The electric potential satisfies

$$
\nabla^2\phi(\mathbf r,t)=-\frac{\rho(\mathbf r,t)}{\varepsilon_0}, \tag{0.2}
\qquad 
\mathbf E=-\nabla \phi .
$$

A homogeneous plasma with $n_e=n_i=n_0$ produces no macroscopic electric field. Neutrality is a physical requirement; in the absence of a compensating positive background, an electron cloud carries net space charge, generates a macroscopic self-field, and does not admit a spatially uniform equilibrium state about which small electrostatic oscillations may be consistently linearized.

The first intrinsic plasma length follows from the static shielding problem. Take the ions as fixed on the electron response time, with $n_i=n_0$, and perturb the electrons by a small electrostatic potential $\phi$. In a thermalized electron population at temperature $T_e$, the electron density follows the Boltzmann factor associated with electron potential energy $-e\phi$:

$$
n_e=n_0\exp\!\left(\frac{e\phi}{k_B T_e}\right). \tag{0.3}
$$

The weak-field condition $|e\phi|\ll k_B T_e$ gives

$$
n_e
=n_0\left[1+\frac{e\phi}{k_B T_e}
+O\!\left(\frac{e^2\phi^2}{k_B^2T_e^2}\right)\right]. \tag{0.4}
$$

To first order,

$$
\rho
=e(n_0-n_e)
\simeq
-\frac{n_0 e^2}{k_B T_e}\phi . \tag{0.5}
$$

Substitution into Poisson's equation gives

$$
\nabla^2\phi
=\frac{n_0 e^2}{\varepsilon_0 k_B T_e}\phi . \tag{0.6}
$$

Writing the coefficient as $1/\lambda_D^2$ defines the Debye length,

$$
\boxed{
\lambda_D=
\left(\frac{\varepsilon_0 k_B T_e}{n_0e^2}\right)^{1/2} \tag{0.7}
}
$$

so the linearized shielding equation becomes

$$
\nabla^2\phi-\frac{\phi}{\lambda_D^2}=0 \tag{0.8}
$$

away from the inserted test charge. With a point charge $Q$ at the origin, the solution is the screened Coulomb potential

$$
\phi(r)=\frac{Q}{4\pi\varepsilon_0 r}\exp\!\left(-\frac{r}{\lambda_D}\right). \tag{0.9}
$$

The formula shows the balance built into $\lambda_D$. Higher temperature weakens screening because thermal motion resists electron accumulation. Higher density strengthens screening because more charges are available to polarize around the perturbation.

The same length scale determines whether the collective description is statistically meaningful. Screening is produced by particles contained in a Debye sphere. The expected number of particles in that volume is

$$
N_D=\frac{4\pi}{3}n_0\lambda_D^3 . \tag{0.10}
$$

Many plasma texts use the order-of-magnitude plasma parameter

$$
\boxed{\Lambda=n_0\lambda_D^3} \tag{0.11}
$$

and absorb $4\pi/3$ into convention. The condition

$$
\Lambda\gg1 \tag{0.12}
$$

is stronger than the statement that the gas is ionized. It states that shielding is a many-particle effect. Relative number fluctuations inside a Debye sphere scale as

$$
\frac{\delta N_D}{N_D}\sim\frac{1}{\sqrt{N_D}}, \tag{0.13}
$$

so the smooth field becomes well defined when $N_D$ is large. The same ordering is also weak coupling. The Coulomb energy of two elementary charges separated by $\lambda_D$ is

$$
U_D=\frac{e^2}{4\pi\varepsilon_0\lambda_D}, \tag{0.14}
$$

and division by the thermal energy gives

$$
\frac{U_D}{k_BT_e}
=
\frac{e^2}{4\pi\varepsilon_0\lambda_D k_BT_e}
=
\frac{1}{4\pi n_0\lambda_D^3}
=
\frac{1}{4\pi\Lambda}. \tag{0.15}
$$

Thus $\Lambda\gg1$ is equivalent, up to constants, to weak electrostatic coupling at the screening scale. Individual close encounters remain present, but they are not the leading mechanism over the electron collective time.

The corresponding time scale is the electron plasma frequency. If the electron fluid is displaced a small distance $\xi$ relative to a fixed ion background, the separation creates charge sheets. Gauss's law gives the restoring field magnitude

$$
E=\frac{e n_0 \xi}{\varepsilon_0}. \tag{0.17}
$$

The electron equation of motion is

$$
m_e\frac{d^2\xi}{dt^2}=-eE
=-\frac{n_0e^2}{\varepsilon_0}\xi . \tag{0.18}
$$

Hence

$$
\frac{d^2\xi}{dt^2}+\omega_{pe}^2\xi=0,
\qquad
\boxed{
\omega_{pe}=
\left(\frac{n_0e^2}{\varepsilon_0 m_e}\right)^{1/2}
}. \tag{0.19}
$$

The ion background is immobile on this time scale to leading order because $m_i\gg m_e$, so the ion plasma frequency is smaller by $(m_e/m_i)^{1/2}$. The neutral plasma specified in the project is therefore the equilibrium state in which fast electron oscillations have a restoring force but no net space-charge acceleration.

The Vlasov ordering used in the following sections is consequently

$$
L\gg\lambda_D,\qquad
\Lambda\gg1,\qquad
\nu_{\mathrm{coll}}\ll\omega_{pe}, \tag{0.20}
$$

where $L$ is a macroscopic length and $\nu_{\mathrm{coll}}$ is a collisional relaxation rate. Under this ordering the plasma remains quasineutral on large scales, but small deviations from neutrality generate the electrostatic fields responsible for Langmuir waves.

Figure 1 below illustrates how $\lambda_D$, $\Lambda$, and $\omega_{pe}$ vary jointly with electron density at fixed temperature $T_e = 10\,\mathrm{eV}$, confirming that the operating point of any realistic Langmuir-wave simulation lies comfortably within this ordering. With the relevant length and time scales in hand, the next step is to specify the mathematical object that will represent the electron population during those oscillations. Because the damping mechanism studied in the later sections depends entirely on how many particles travel at each speed, a fluid description that retains only density and mean velocity is insufficient. Section 1 introduces the phase-space distribution function, the kinetic object that preserves this velocity-space information.

````python
Te_eV = 10.0
Te = Te_eV * 11604.51812
n0 = np.logspace(14, 22, 700)

lambda_D = np.sqrt(eps0 * kB * Te / (n0 * e**2))
Lambda = n0 * lambda_D**3
omega_pe = np.sqrt(n0 * e**2 / (eps0 * me))

fig, axes = plt.subplots(1, 3, figsize=(13.5, 3.6))

axes[0].loglog(n0, lambda_D)
axes[0].set_xlabel(r"$n_0\;(\mathrm{m^{-3}})$")
axes[0].set_ylabel(r"$\lambda_D\;(\mathrm{m})$")
axes[0].set_title(r"Debye length, $T_e=10\,\mathrm{eV}$")
axes[0].grid(True, which="both", alpha=0.25)

axes[1].loglog(n0, Lambda)
axes[1].axhline(1, linestyle="--", linewidth=1)
axes[1].set_xlabel(r"$n_0\;(\mathrm{m^{-3}})$")
axes[1].set_ylabel(r"$\Lambda=n_0\lambda_D^3$")
axes[1].set_title("Plasma parameter")
axes[1].grid(True, which="both", alpha=0.25)

axes[2].loglog(n0, omega_pe)
axes[2].set_xlabel(r"$n_0\;(\mathrm{m^{-3}})$")
axes[2].set_ylabel(r"$\omega_{pe}\;(\mathrm{s^{-1}})$")
axes[2].set_title("Electron plasma frequency")
axes[2].grid(True, which="both", alpha=0.25)

fig.suptitle("Figure 1. Screening length, collective particle number, and electron time scale")
fig.tight_layout()
plt.show()

n_check = 1e18
lambda_check = np.sqrt(eps0 * kB * Te / (n_check * e**2))
Lambda_check = n_check * lambda_check**3
omega_check = np.sqrt(n_check * e**2 / (eps0 * me))

print(f"At n0 = {n_check:.1e} m^-3 and Te = {Te_eV:.1f} eV:")
print(f"lambda_D = {lambda_check:.3e} m")
print(f"Lambda   = {Lambda_check:.3e}")
print(f"omega_pe = {omega_check:.3e} s^-1")
````

**Figure 1. Screening length, collective particle number, and electron time scale**

Figure 1 summarizes the density dependence of the principal electrostatic scales at fixed electron temperature $T_e=10\,\mathrm{eV}$. Only relations needed for later kinetic modeling are emphasized.

The Debye length decreases with increasing density,

$$
\lambda_D=\left(\frac{\varepsilon_0 k_B T_e}{n_0 e^2}\right)^{1/2}, 
$$

so denser plasmas screen charge perturbations over shorter distances. This identifies the spatial scale below which charge separation may persist and above which quasineutrality is rapidly restored.

The plasma parameter,

$$
\Lambda=n_0\lambda_D^3, 
$$

also decreases as $n_0^{-1/2}$ at fixed temperature, yet remains large throughout the plotted range. A large value of $\Lambda$ indicates that many particles participate in shielding, which supports the mean-field ordering required for the Vlasov model.

The electron plasma frequency increases with density,

$$
\omega_{pe}=\left(\frac{n_0 e^2}{\varepsilon_0 m_e}\right)^{1/2},
$$

and sets the characteristic timescale for rapid electron response. Higher density therefore corresponds to faster Langmuir-scale oscillations.

At the representative value $n_0=10^{18}\,\mathrm{m^{-3}}$,

$$
\lambda_D=2.35\times10^{-5}\,\mathrm{m},\qquad
\Lambda=1.30\times10^4,\qquad
\omega_{pe}=5.64\times10^{10}\,\mathrm{s^{-1}}. 
$$

These values place the system in a strongly collective and weakly coupled regime appropriate for collisionless kinetic analysis.

The three curves together encode the ordering $\Lambda\gg1$ and $\omega_{pe}\gg\nu_\text{coll}$ that legitimises the mean-field treatment. What they do not yet specify is the mathematical object used to represent the electron population within that ordering. Knowing that the plasma is collective does not by itself say how particle speeds are distributed, yet the wave-particle resonance responsible for Landau damping depends precisely on that distribution near a single velocity — the wave phase speed. Section 1 therefore introduces the distribution function as the minimal kinetic object that retains this velocity-space information.

## 1. Distribution Function in Phase Space


Section 0 established the length and time scales that characterise collective plasma behaviour; it did not, however, specify how particle velocities are distributed. That information is encoded in the phase-space distribution function introduced here.

The ordering above removes the need to follow individual Coulomb encounters, but it does not justify a fluid model by itself. Langmuir waves and collisionless damping depend on how particles are distributed in velocity space. A description retaining velocity information is therefore required.

For one given particle, the kinetic state is represented by a distribution function $f(\mathbf r,\mathbf v,t)$ on the six-dimensional phase space. It is defined by

$$
dN=f(\mathbf r,\mathbf v,t)\,d^3r\,d^3v, \tag{1.1}
$$

so that $dN$ is the number of particles in a spatial element $d^3r$ centered at $\mathbf r$ and a velocity-space element $d^3v$ centered at $\mathbf v$. Since $d^3v$ has units $\mathrm{m^3s^{-3}}$, the dimensions of $f$ are

$$
[f]=\mathrm{m^{-3}}(\mathrm{m\,s^{-1}})^{-3}
=\mathrm{s^3m^{-6}} . \tag{1.2}
$$

In kinetic theory, $\mathbf r$ and $\mathbf v$ are independent variables. A single particle trajectory has $\mathbf r=\mathbf r(t)$ and $\mathbf v=\dot{\mathbf r}(t)$; a distribution function is instead a field over all possible positions and velocities at a given time. This distinction is essential because acceleration changes the occupancy of velocity space, not only the density in physical space.

The connection with fluid quantities is made through moments. The number density is

$$
n(\mathbf r,t)=\int_{\mathbb R^3} f(\mathbf r,\mathbf v,t)\,d^3v \tag{1.3}
$$

The mean flow velocity is

$$
\mathbf u(\mathbf r,t)
=
\frac{1}{n(\mathbf r,t)}
\int_{\mathbb R^3}\mathbf v f(\mathbf r,\mathbf v,t)\,d^3v. \tag{1.4}
$$

The pressure tensor is

$$
P_{ij}(\mathbf r,t)
=
m\int_{\mathbb R^3}
\big(v_i-u_i\big)\big(v_j-u_j\big)
f(\mathbf r,\mathbf v,t)\,d^3v . \tag{1.5}
$$

Fluid theory retains a finite number of these moments and then closes the hierarchy by assumption. Kinetic theory keeps the distribution itself, and therefore retains the information needed to describe velocity-space resonances and non-Maxwellian structure.

For a spatially homogeneous electron equilibrium with no drift, the appropriate background distribution is the Maxwellian

$$
f_0(\mathbf v)
=
n_0
\left(\frac{m_e}{2\pi k_B T_e}\right)^{3/2}
\exp\!\left(-\frac{m_e|\mathbf v|^2}{2k_B T_e}\right). \tag{1.6}
$$

The normalization follows from the Gaussian integral

$$
\int_{-\infty}^{\infty}
\exp\!\left(-\frac{m_ev_x^2}{2k_BT_e}\right)\,dv_x
=
\left(\frac{2\pi k_BT_e}{m_e}\right)^{1/2}. \tag{1.6}
$$

Since the three velocity components factor,

$$
\int_{\mathbb R^3} f_0(\mathbf v)\,d^3v
=
n_0
\left(\frac{m_e}{2\pi k_B T_e}\right)^{3/2}
\left(\frac{2\pi k_B T_e}{m_e}\right)^{3/2}
=
n_0. \tag{1.7}
$$

The Maxwellian may also be obtained variationally. Maximizing the kinetic entropy

$$
S[f]=-k_B\int f\ln f\,d^3v \tag{1.8}
$$

subject to fixed particle density

$$
\int f\,d^3v=n_0 \tag{1.9}
$$

and fixed kinetic energy density

$$
\int \frac12 m_ev^2 f\,d^3v=\frac32 n_0k_BT_e \tag{1.10}
$$

gives

$$
\delta\left[
-k_B\int f\ln f\,d^3v
-\alpha\int f\,d^3v
-\beta\int \frac12m_ev^2f\,d^3v
\right]=0. \tag{1.11}
$$

The coefficient of $\delta f$ must vanish:

$$
-k_B(\ln f+1)-\alpha-\frac{\beta m_ev^2}{2}=0. \tag{1.12}
$$

Thus

$$
f=A\exp(-Bv^2), \tag{1.13}
$$

with constants fixed by the two constraints. This gives the Maxwellian above with $B=m_e/(2k_BT_e)$. The equilibrium is uniform in space, isotropic in velocity, and stationary in time. It is the background distribution from which the small-amplitude electron density perturbations of Langmuir theory are constructed.

Figure 2 below plots this equilibrium in one-dimensional velocity space and in the full $(x,v)$ phase plane, making explicit that $f_0$ carries no spatial dependence and is symmetric about $v=0$ — the precise state around which the Langmuir perturbation will be superimposed. The smooth function $f_0$ has been introduced here as a postulate, defined by its physical properties rather than derived. Section 2 supplies the derivation: it traces the formal path from the exact, singular Klimontovich density of $N$ point particles to this smooth mean-field object, and identifies the large-$\Lambda$ condition as the step that permits the binary-collision correlations to be consistently discarded.

````python
vT = 1.0
v = np.linspace(-5*vT, 5*vT, 800)
f1 = np.exp(-v**2/(2*vT**2)) / np.sqrt(2*np.pi*vT**2)

x = np.linspace(0, 1, 320)
X, V = np.meshgrid(x, v)
F = np.exp(-V**2/(2*vT**2)) / np.sqrt(2*np.pi*vT**2)

fig, axes = plt.subplots(1, 2, figsize=(10.5, 3.8))

axes[0].plot(v/vT, f1)
axes[0].set_xlabel(r"$v/v_T$")
axes[0].set_ylabel(r"$f_0(v)$")
axes[0].set_title("One-dimensional Maxwellian")
axes[0].grid(alpha=0.25)

mesh = axes[1].pcolormesh(X, V/vT, F, shading="auto")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$v/v_T$")
axes[1].set_title("Homogeneous equilibrium in phase space")
fig.colorbar(mesh, ax=axes[1], label=r"$f_0$")

fig.suptitle("Figure 2. Maxwellian equilibrium: velocity structure without spatial variation")
fig.tight_layout()
plt.show()
````

**Figure 2. Maxwellian equilibrium in velocity and phase space**

Figure 2 presents the equilibrium state used in the linearized Vlasov analysis.

The left panel shows the one-dimensional Maxwellian velocity distribution,

$$
f_0(v)=\frac{1}{\sqrt{2\pi}v_T}\exp\!\left(-\frac{v^2}{2v_T^2}\right),
$$

centered at zero velocity. The symmetry about $v=0$ indicates zero bulk drift, while the width is determined by the thermal speed $v_T$.

The right panel shows the same equilibrium in $(x,v)$ phase space. Because the state is spatially homogeneous, the distribution is independent of position and varies only with velocity. This provides the uniform background from which small electrostatic perturbations are introduced.

Both panels confirm that $f_0$ is smooth, spatially uniform, and symmetric in velocity, properties that follow from thermal equilibrium and the quasineutrality condition established in Section 0. What the figure does not show is how this smooth object emerges from the underlying reality of $N$ discrete, interacting point charges. Section 2 closes that gap by averaging the exact Klimontovich equation, making explicit which terms are kept and which are dropped, and under what conditions the resulting Vlasov–Poisson system faithfully replaces the $N$-body problem.

## 2. From Exact Particle Dynamics to the Vlasov Limit

Section 1 postulated $f$ as a smooth density on phase space. The question left open is whether this object can be derived systematically from the exact $N$-particle description, and what approximation is required to close the equation it satisfies. That derivation begins here.

The distribution function used in the Vlasov equation is the smooth limit of an exact microscopic density. For $N$ particles of a given species, with trajectories $\mathbf r_k(t)$ and velocities $\mathbf v_k(t)$, define the Klimontovich density

$$
\mathcal N(\mathbf r,\mathbf v,t)
=
\sum_{k=1}^{N}
\delta^{(3)}\!\big(\mathbf r-\mathbf r_k(t)\big)
\delta^{(3)}\!\big(\mathbf v-\mathbf v_k(t)\big). \tag{2.1}
$$

It is singular rather than smooth. Each particle contributes one spike at its instantaneous phase-space location. Integration over any phase-space region counts exactly how many particles lie in that region.

The equation for $\mathcal N$ follows directly from the chain rule. For one particle,

$$
\frac{\partial}{\partial t}
\delta^{(3)}(\mathbf r-\mathbf r_k(t))
=
-\dot{\mathbf r}_k(t)\cdot
\nabla_r\delta^{(3)}(\mathbf r-\mathbf r_k(t)),
$$

and

$$
\frac{\partial}{\partial t}
\delta^{(3)}(\mathbf v-\mathbf v_k(t))
=
-\dot{\mathbf v}_k(t)\cdot
\nabla_v\delta^{(3)}(\mathbf v-\mathbf v_k(t)).
$$

Applying the product rule to the two delta factors and summing over $k$,

$$
\frac{\partial \mathcal N}{\partial t}
=
-\sum_k
\dot{\mathbf r}_k\cdot\nabla_r
\left[
\delta^{(3)}(\mathbf r-\mathbf r_k)
\delta^{(3)}(\mathbf v-\mathbf v_k)
\right]
-\sum_k
\dot{\mathbf v}_k\cdot\nabla_v
\left[
\delta^{(3)}(\mathbf r-\mathbf r_k)
\delta^{(3)}(\mathbf v-\mathbf v_k)
\right]. \tag{2.2}
$$

On the support of the $k$-th delta function, $\mathbf v=\mathbf v_k$. Newton's equation gives

$$
\dot{\mathbf v}_k=\frac{\mathbf F_{\mathrm{exact}}(\mathbf r_k,t)}{m}.
$$

The exact microscopic density therefore satisfies

$$
\boxed{
\frac{\partial \mathcal N}{\partial t}
+\mathbf v\cdot\nabla_r\mathcal N
+\frac{\mathbf F_{\mathrm{exact}}}{m}\cdot\nabla_v\mathcal N
=0
} \tag{2.3}
$$

in the distributional sense. This is the Klimontovich equation: a Newtonian mechanics in phase-space conservation.

The smooth kinetic distribution is introduced by ensemble averaging or by coarse graining over volumes large compared with inter-particle spacing but small compared with macroscopic scales:

$$
f(\mathbf r,\mathbf v,t)=\langle \mathcal N(\mathbf r,\mathbf v,t)\rangle.
$$

Write the microscopic electric field and density as mean plus fluctuation,

$$
\mathbf E_{\mathrm{exact}}=\mathbf E+\delta\mathbf E,
\qquad
\mathcal N=f+\delta\mathcal N. \tag{2.4}
$$

Averaging the Klimontovich equation gives, for the electrostatic force $q\mathbf E_{\mathrm{exact}}$,

$$
\frac{\partial f}{\partial t}
+\mathbf v\cdot\nabla_r f
+\frac{q}{m}\mathbf E\cdot\nabla_v f
=
-\frac{q}{m}
\left\langle
\delta\mathbf E\cdot\nabla_v\delta\mathcal N
\right\rangle . \tag{2.5}
$$

The term on the right contains the correlation between microscopic field fluctuations and microscopic density fluctuations. Under weak-coupling closures it generates collisional kinetic operators, including Landau or Balescu--Lenard forms. It is absent only in the collisionless mean-field approximation.

The ordering $\Lambda\gg1$ supplies the approximation. Since many particles occupy a Debye sphere, fluctuations in the microscopic field are small compared with the averaged field at the collective scale. Collisional relaxation is slower than the electron plasma period, written schematically as

$$
\nu_{\mathrm{coll}}\ll\omega_{pe}.
$$

Neglecting the fluctuation-correlation term gives the Vlasov equation:

$$
\boxed{
\frac{\partial f}{\partial t}
+\mathbf v\cdot\nabla_r f
+\frac{q}{m}\mathbf E\cdot\nabla_v f=0
}. \tag{2.6}
$$

The electric field is generated by the same distribution that it accelerates. For an electron distribution $f_e$ and fixed ion background $n_i=n_0$,

$$
\rho=e n_0-e\int f_e\,d^3v,
$$

so Poisson's equation gives

$$
\nabla\cdot\mathbf E=\frac{1}{\varepsilon_0}
\left(
e n_0-e\int f_e\,d^3v
\right),
\qquad
\mathbf E=-\nabla\phi.
$$

Equivalently,

$$
-\nabla^2\phi
=
\frac{e}{\varepsilon_0}
\left(
n_0-\int f_e\,d^3v
\right).
$$

The Vlasov equation together with this Poisson equation forms the Vlasov--Poisson system used for electrostatic Langmuir dynamics.

The derivation above answers the question of *what equation* governs $f$. The equation has been obtained; it remains to ask what *geometric structure* that equation possesses. Specifically: is the phase-space flow generated by the Lorentz force compressible, and what conservation laws follow from the answer? These questions matter not only for the physics — they determine whether the plasma can truly forget its initial state — but also for numerics, since a solver that violates the resulting invariants introduces unphysical damping or growth that can mask the Landau signal the simulation is designed to observe. Section 3 addresses both points through Liouville's theorem.

## 3. Liouville Structure and Collisionless Transport

The Vlasov equation derived in Section 2 is a first-order advection equation in six-dimensional phase space. Whether its characteristics preserve phase-space volume — and whether fine-grained density along those characteristics is conserved — depends on the divergence of the advecting flow, which is computed here.

Collisionless motion transports phase-space density along the characteristics generated by the Lorentz force.

For a particle of charge $q$ and mass $m$,

$$
\dot{\mathbf r}=\mathbf v,
\qquad
\dot{\mathbf v}
=
\frac{q}{m}
\left(
\mathbf E+\mathbf v\times\mathbf B
\right)
$$

with the six-dimensional coordinate:

$$
\mathbf w=(\mathbf r,\mathbf v),
\qquad
\dot{\mathbf w}=(\dot{\mathbf r},\dot{\mathbf v}).
$$

the phase-space divergence of this flow is

$$
\nabla_w\cdot\dot{\mathbf w}
=
\nabla_r\cdot\mathbf v
+
\nabla_v\cdot
\frac{q}{m}
\left(
\mathbf E+\mathbf v\times\mathbf B
\right).
$$

The first term vanishes because velocity is independent of the position coordinates:

$$
\nabla_r\cdot\mathbf v
=
\sum_{i=1}^{3}\frac{\partial v_i}{\partial r_i}
=0.
$$

The electric part of the second term vanishes because $\mathbf E=\mathbf E(\mathbf r,t)$ has no velocity dependence:

$$
\nabla_v\cdot\mathbf E
=
\sum_{i=1}^{3}\frac{\partial E_i}{\partial v_i}
=0.
$$

For the magnetic term, use $(\mathbf v\times\mathbf B)_i=\varepsilon_{ijk}v_jB_k$. Then

$$
\nabla_v\cdot(\mathbf v\times\mathbf B)
=
\frac{\partial}{\partial v_i}
\left(\varepsilon_{ijk}v_jB_k\right)
=
\varepsilon_{ijk}\delta_{ij}B_k
=
\varepsilon_{jjk}B_k
=0.
$$

Thus

$$
\boxed{
\nabla_w\cdot\dot{\mathbf w}=0
}.
$$

This is Liouville's theorem for the nonrelativistic charged-particle flow: phase-space volume is preserved under Hamiltonian motion. A material element in phase space can stretch and shear, though not compressed.

Particle conservation in phase space is expressed by the continuity equation

$$
\frac{\partial f}{\partial t}
+
\nabla_w\cdot(f\dot{\mathbf w})=0 .
$$

Expanding the divergence,

$$
\frac{\partial f}{\partial t}
+
\dot{\mathbf w}\cdot\nabla_w f
+
f\,\nabla_w\cdot\dot{\mathbf w}
=0.
$$

Liouville's theorem removes the last term, leaving

$$
\frac{\partial f}{\partial t}
+
\dot{\mathbf w}\cdot\nabla_w f=0.
$$

Substituting the equations of motion gives the full collisionless Vlasov equation,

$$
\boxed{
\frac{\partial f}{\partial t}
+
\mathbf v\cdot\nabla_r f
+
\frac{q}{m}
(\mathbf E+\mathbf v\times\mathbf B)\cdot\nabla_v f
=0
}.
$$

For the unmagnetized electrostatic electron model used here,

$$
\mathbf B=0,\qquad
\mathbf E=-\nabla\phi,\qquad
q=-e,
$$

so the electron Vlasov equation becomes

$$
\boxed{
\frac{\partial f}{\partial t}
+
\mathbf v\cdot\nabla_r f
-
\frac{e}{m_e}\mathbf E\cdot\nabla_v f
=0
}.
$$

Along a characteristic curve $(\mathbf r(t),\mathbf v(t))$ satisfying

$$
\frac{d\mathbf r}{dt}=\mathbf v,
\qquad
\frac{d\mathbf v}{dt}=-\frac{e}{m_e}\mathbf E(\mathbf r,t),
$$

the total derivative of the distribution is

$$
\frac{df}{dt}
=
\frac{\partial f}{\partial t}
+
\frac{d\mathbf r}{dt}\cdot\nabla_r f
+
\frac{d\mathbf v}{dt}\cdot\nabla_v f
=0.
$$

The fine-grained phase-space density is conserved along particle characteristics. Although phase mixing may generate increasingly fine filamentary structure, collisionless Vlasov evolution alone does not produce irreversible relaxation toward Maxwellian thermodynamic equilibrium.

This also gives a useful invariant. For any sufficiently smooth function $H$,

$$
\frac{d}{dt}\int H(f)\,d^3r\,d^3v
=
\int H'(f)\frac{\partial f}{\partial t}\,d^3r\,d^3v.
$$

Using the conservative form of the Vlasov equation,

$$
\frac{\partial f}{\partial t}=-\nabla_w\cdot(f\dot{\mathbf w}),
$$

and assuming periodic boundaries or decay at infinity, integration by parts gives

$$
\frac{d}{dt}\int H(f)\,d^3r\,d^3v=0.
$$

The Boltzmann functional

$$
S=-k_B\int f\ln f\,d^3r\,d^3v
$$

is therefore conserved by the fine-grained Vlasov equation. Numerical methods that introduce artificial diffusion change this invariant structure. For a simulation intended to demonstrate Langmuir waves, preserving mass and avoiding unphysical damping are not merely numerical preferences; they follow from the collisionless theory itself.

Figure 3 below makes this concrete by showing the simplest collisionless transport scenario — free streaming in the absence of any applied field — and illustrating how an initially localised phase-space element shears without compressing, providing a geometric realisation of the zero-divergence condition proved above.


````python
np.random.seed(4)

N = 6000
x0 = np.random.normal(0.0, 0.16, N)
v0 = np.random.normal(0.0, 0.30, N)

times = [0.0, 2.0, 5.0]
fig, axes = plt.subplots(1, 3, figsize=(12, 3.6), sharex=True, sharey=True)

for ax, t in zip(axes, times):
    x = x0 + v0 * t
    ax.scatter(x, v0, s=0.45, alpha=0.35, rasterized=True)
    ax.set_title(rf"$t={t:g}$")
    ax.set_xlabel(r"$x$")
    ax.grid(alpha=0.2)

axes[0].set_ylabel(r"$v$")
axes[0].set_xlim(-3.0, 3.0)
axes[0].set_ylim(-1.25, 1.25)

fig.suptitle("Figure 3. Free streaming shears a phase-space cloud without changing its fine-grained density")
fig.tight_layout()
plt.show()
````

**Figure 3. Free streaming and phase-space shear**

Figure 3 illustrates collisionless transport in the absence of forces,

$$
\frac{dx}{dt}=v,\qquad \frac{dv}{dt}=0.
$$

Particle velocities remain constant, while positions evolve according to

$$
x(t)=x_0+vt.
$$

The initially localized cloud is therefore sheared in phase space: particles with larger positive velocity move to increasing $x$, while those with negative velocity move oppositely.

The occupied region changes shape but not fine-grained phase-space area. This is the transport geometry underlying Liouville's theorem and the collisionless Vlasov description.

The figure also indicates why numerical resolution is important. Smooth initial data develop increasingly fine filamentary structure under advection, so excessive numerical diffusion can suppress physically relevant phase-space evolution.

Section 2 derived the Vlasov equation by averaging the Klimontovich description under the mean-field ordering $\Lambda\gg1$. Section 3 demonstrated that the resulting equation is a volume-preserving advection in phase space, generating a hierarchy of exact conservation laws that any faithful solver must respect. ## 6. The Climax: Langmuir Waves, the Landau Prescription, and Collisionless Damping


In Section 5, we linearised the Vlasov–Poisson system around a spatially uniform, time-steady equilibrium $f_0(\mathbf{v})$. By assuming plane-wave perturbations of the form $f_1, \phi_1 \propto e^{i(kx - \omega t)}$, we converted the coupled partial differential equations into an algebraic condition — the **electrostatic dispersion relation**:

$$\boxed{\varepsilon(k, \omega) \equiv 1 - \frac{\omega_{pe}^2}{k^2} \int_{-\infty}^{\infty} \frac{\partial \hat{f}_0/\partial v}{v - \omega/k}\,dv = 0,} \tag{5.8}$$

where $\hat{f}_0 = f_0/n_0$ is the normalised equilibrium distribution. This condition encodes the "rulebook" of the plasma: only frequency–wavenumber pairs $(\omega, k)$ that satisfy $\varepsilon = 0$ correspond to physically self-sustaining waves. We called this the **Dielectric Function** — a measure of the plasma's willingness to support an oscillation without any external driver.

However, Section 5 left two critical questions unanswered:

1. **The singularity problem.** The integrand in (5.8) diverges when $v = \omega/k$. A wave with phase velocity $\omega/k$ resonates with particles moving at that exact speed — and naive calculus breaks down here. *How do we evaluate this integral?*

2. **The dispersion answer.** Even if we can evaluate the integral, what are the actual solutions $\omega(k)$? What do they tell us about the frequency and lifetime of the plasma waves?

Section 6 answers both questions in sequence. Subsection 6.1 resolves the singularity through the **Landau Prescription** — a mathematically precise detour into the complex plane that captures the physics of resonant particles. Subsection 6.2 uses the resolved integral to derive the **Bohm–Gross Dispersion Relation**, giving the real oscillation frequency $\omega_r(k)$. Subsection 6.3 extracts the imaginary part to obtain the **Landau Damping Rate** $\gamma(k)$, the collisionless decay mechanism that is the central physical prediction of the entire theory.

---

## 6.1 The Landau Prescription: Resolving the Resonant Denominator

### 6.1.1 Background: Why the Singularity is Not a Failure

Before deriving the resolution, it is worth understanding why the singularity at $v = \omega/k$ exists and what it physically represents. This is not a flaw in the mathematics — it is the mathematics trying to tell us something important.

**The phase velocity** $v_\phi = \omega/k$ is the speed at which the wave's crests travel through the plasma. A particle moving at exactly this speed *sees a static electric field*. It does not experience an oscillating push-and-pull; from its reference frame, the wave is frozen. This is the condition of **resonance**: the particle and the wave are locked in step.

In the frame of such a particle, even a very weak field can do sustained work — it can continuously accelerate or decelerate the particle, transferring energy between the wave and the particle population. This sustained energy exchange is precisely what the singularity at $v = \omega/k$ is encoding. The denominator going to zero means the effect of the wave on resonant particles is *not small*; it is of a qualitatively different character from its effect on non-resonant ones.

The mathematical challenge is: how do we integrate through this infinity in a way that faithfully represents the physics?

### 6.1.2 The Root Cause: The Normal Mode Assumption

The difficulty originates from an implicit assumption made in Section 5. When we wrote $f_1 \propto e^{-i\omega t}$ and assumed $\omega$ to be a **real** number, we were treating the wave as a **perfect, eternal oscillation** — one that has existed since $t = -\infty$ with constant amplitude. This is the "Normal Mode" assumption, and it is physically artificial.

In any real experiment, the wave is **turned on** at some finite time $t = 0$. Before $t = 0$, the plasma is undisturbed; after $t = 0$, it responds. This is an **Initial Value Problem**, not a normal mode problem. Landau (1946) was the first to recognise that treating it correctly as an initial value problem was the key to resolving the singularity.

The appropriate mathematical tool is the **Laplace Transform** in time (rather than the Fourier Transform). The Laplace transform variable $s = \gamma + i\omega_r$ has a positive real part during the inversion, which means $\omega = \omega_r + i\gamma$ with $\gamma > 0$ initially. The singularity $v = \omega/k$ is therefore displaced off the real $v$-axis into the complex plane:

$$v_\text{singular} = \frac{\omega}{k} = \frac{\omega_r}{k} + i\frac{\gamma}{k}.$$

Because $\gamma > 0$, this point lies in the **upper half** of the complex $v$-plane — above the real axis. The integration contour (along the real $v$-axis) therefore passes *below* the singularity, without touching it. The integral is well-defined.

The physical question is: what happens when we let $\gamma \to 0^+$ (i.e., we ask about the long-time behaviour, where the system has settled into something close to a steady oscillation)? The singularity descends toward the real axis from above, and we must track how the integral changes as it does so. This limiting process is the essence of the Landau prescription.

### 6.1.3 The Landau Contour

The prescription can be stated geometrically. Let $a = \omega_r/k$ be the real phase velocity. For $\gamma > 0$, the singularity is at $v = a + i(\gamma/k)$, above the real axis, and the contour $\mathcal{C}$ is simply the real line. As $\gamma \to 0^+$:

- The singularity approaches the real axis from above.
- To maintain *analyticity* (i.e., to ensure the solution is causal — the plasma responds **after** the wave starts, not before), the contour must be deformed to pass **below** the singularity.

This deformed contour is the **Landau contour** $\mathcal{C}_L$. It dips below the point $v = a$ on the real axis:

```
       ──────────────────────────────── Im(v) = 0
                            ↙
               (contour dips below singularity here)
```

Any other choice of contour would correspond to an **acausal** solution — one where the plasma responds before the wave starts. Causality therefore uniquely fixes the integration path.

### 6.1.4 Derivation of the Plemelj–Sokhotski Identity

We now derive the mathematical consequence of the Landau prescription. We want to evaluate the limiting integral:

$$I = \lim_{\varepsilon \to 0^+} \int_{-\infty}^{\infty} \frac{g(v)}{v - a - i\varepsilon}\,dv,$$

where $a = \omega_r/k$ is real and $\varepsilon = \gamma/k > 0$ is small, and $g(v) = \partial \hat{f}_0/\partial v$ is a smooth, rapidly decaying function.

**Step 1: Rationalise the denominator.**

Multiply numerator and denominator by the complex conjugate $(v - a + i\varepsilon)$:

$$\frac{1}{v - a - i\varepsilon} = \frac{(v-a) + i\varepsilon}{(v-a)^2 + \varepsilon^2}.$$

The integral splits into a real part and an imaginary part:

$$I = \int_{-\infty}^{\infty} g(v)\,\frac{v - a}{(v-a)^2 + \varepsilon^2}\,dv \;+\; i\varepsilon \int_{-\infty}^{\infty} g(v)\,\frac{1}{(v-a)^2 + \varepsilon^2}\,dv. \tag{6.1a}$$

**Step 2: Evaluate the imaginary part as $\varepsilon \to 0^+$.**

The function $\varepsilon / [(v-a)^2 + \varepsilon^2]$ is a **Lorentzian** of width $\varepsilon$ centred at $v = a$. In the limit $\varepsilon \to 0^+$, it concentrates into a Dirac delta:

$$\lim_{\varepsilon \to 0^+} \frac{\varepsilon}{(v-a)^2 + \varepsilon^2} = \pi\,\delta(v - a).$$

This is a standard result: the Lorentzian is a delta-sequence. (It can be verified by checking that $\int_{-\infty}^\infty \varepsilon/[(v-a)^2+\varepsilon^2]\,dv = \pi$ for all $\varepsilon > 0$.) Therefore, the imaginary part of $I$ becomes:

$$\lim_{\varepsilon \to 0^+} i\varepsilon \int_{-\infty}^{\infty} \frac{g(v)}{(v-a)^2 + \varepsilon^2}\,dv = i\pi\int_{-\infty}^{\infty} g(v)\,\delta(v-a)\,dv = i\pi\,g(a). \tag{6.1b}$$

**Step 3: Evaluate the real part as $\varepsilon \to 0^+$.**

The function $(v-a)/[(v-a)^2+\varepsilon^2]$ is an odd function of $(v-a)$. Near $v = a$, the integrand is approximately $(v-a)/\varepsilon^2 \cdot g(v)$, which is integrable provided we treat the $v=a$ point symmetrically. In the limit $\varepsilon \to 0^+$, this converges to the **Cauchy Principal Value**:

$$\lim_{\varepsilon \to 0^+} \int_{-\infty}^{\infty} g(v)\,\frac{v-a}{(v-a)^2 + \varepsilon^2}\,dv = \mathcal{P}\int_{-\infty}^{\infty} \frac{g(v)}{v - a}\,dv, \tag{6.1c}$$

where the Cauchy principal value is defined as:

$$\mathcal{P}\int_{-\infty}^{\infty} \frac{g(v)}{v - a}\,dv \equiv \lim_{\eta \to 0^+} \left[ \int_{-\infty}^{a-\eta} \frac{g(v)}{v-a}\,dv + \int_{a+\eta}^{\infty} \frac{g(v)}{v-a}\,dv \right].$$

The symmetric exclusion $(-\eta, +\eta)$ around the singularity ensures that the $1/(v-a)$ pole cancels in a pairwise sense — particles slightly slower and slightly faster than the wave each contribute equally and oppositely to the non-resonant response, so their total contribution is finite.

**Step 4: Assemble the Plemelj–Sokhotski identity.**

Combining Steps 2 and 3 with $g(v) = \partial \hat{f}_0 / \partial v$:

$$\boxed{\lim_{\varepsilon \to 0^+} \int_{-\infty}^{\infty} \frac{\partial \hat{f}_0/\partial v}{v - \omega_r/k - i\varepsilon}\,dv = \mathcal{P}\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0/\partial v}{v - \omega_r/k}\,dv + i\pi\left.\frac{\partial \hat{f}_0}{\partial v}\right|_{v=\omega_r/k}.} \tag{6.2}$$

**Variable anatomy:**
- $\mathcal{P}\int$ — the Cauchy Principal Value integral. Represents the **bulk, non-resonant** response of all particles not moving at the wave speed. Physically: the "spring and inertia" of the plasma that determines the wave frequency.
- $i\pi\,\partial \hat{f}_0/\partial v|_{v=\omega_r/k}$ — the **residue term**. Represents the singular contribution of the resonant particles — those particles moving exactly at the wave phase velocity. The imaginary unit $i$ means this term affects the amplitude (growth or decay) of the wave, not its frequency.
- $\omega_r/k$ — the wave phase velocity, $v_\phi$.

### 6.1.5 The Two Physical Parts

The decomposition (6.2) has a direct physical interpretation, which is why it is not just a mathematical trick:

**Part A — The Principal Value (The Bulk Response):** The particles that are *not* in resonance with the wave experience an oscillating electric force that averages to zero over one wave period. They do not exchange net energy with the wave; they simply provide a collective electromagnetic inertia and restoring force that determines *how fast* the wave oscillates. This is analogous to the "mass and spring" of a mechanical oscillator.

**Part B — The Residue (The Resonant Particles):** The particles moving at exactly $v = v_\phi$ are, from the wave's perspective, stationary targets. They experience a constant force for as long as they stay near a wave crest or trough. These "surfers" can exchange energy with the wave continuously. The slope $\partial \hat{f}_0/\partial v$ evaluated at $v_\phi$ determines *which way* that exchange goes:
- If $\partial \hat{f}_0/\partial v < 0$ at $v_\phi$ (i.e., there are more slow particles than fast ones — the normal case for a Maxwellian), the wave loses energy to the slower-moving resonant particles, and the wave decays. This is **Landau Damping**.
- If $\partial \hat{f}_0/\partial v > 0$ at $v_\phi$ (i.e., there are more fast particles than slow — a "bump on tail" instability), the wave gains energy, and the wave grows. This is a kinetic instability.

Substituting (6.2) into the dielectric function (5.8), we can now write:

$$\varepsilon(k, \omega_r) = \underbrace{\varepsilon_R(k, \omega_r)}_{\text{Real part}} + i\underbrace{\varepsilon_I(k, \omega_r)}_{\text{Imaginary part}} = 0, \tag{6.3}$$

where:

$$\varepsilon_R = 1 - \frac{\omega_{pe}^2}{k^2}\,\mathcal{P}\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0/\partial v}{v - \omega_r/k}\,dv, \tag{6.4}$$

$$\varepsilon_I = -\frac{\pi\,\omega_{pe}^2}{k^2}\left.\frac{\partial \hat{f}_0}{\partial v}\right|_{v=\omega_r/k}. \tag{6.5}$$

These two equations — one real, one imaginary — together determine both the oscillation frequency $\omega_r$ and the growth/decay rate $\gamma$. We extract them in the next two subsections.

---

## 6.2 The Bohm–Gross Dispersion Relation

### 6.2.1 Background: Why We Need a Long-Wavelength Expansion

The dielectric condition $\varepsilon_R(k, \omega_r) = 0$ is, in general, a transcendental equation. Even after applying the Landau Prescription to resolve the singularity, the principal value integral involves the full shape of the distribution $f_0$ and has no closed-form solution for arbitrary $\omega_r/k$.

Progress requires identifying the physically relevant regime. For the Langmuir wave, the key observation is:

> **The wave is fast.** Langmuir waves travel at a phase velocity $v_\phi = \omega_r/k \gg v_{Te}$, where $v_{Te} = \sqrt{T_e/m_e}$ is the electron thermal velocity.

This is the **long-wavelength limit** ($k\lambda_D \ll 1$), and it has a precise physical meaning. The Debye length $\lambda_D = v_{Te}/\omega_{pe}$ is the skin depth over which the plasma shields charge perturbations. When $k\lambda_D \ll 1$, the perturbation wavelength $\lambda = 2\pi/k$ is much longer than the shielding length — the plasma has "room" to organise a collective oscillation before shielding destroys it. Equivalently, the ratio $kv_{Te}/\omega_r \sim k\lambda_D \ll 1$, so most thermal electrons move much slower than the wave. The fraction of truly resonant particles (those with $v \approx v_\phi$) is exponentially small.

This smallness is what allows a **Taylor expansion**: we expand the principal value integrand in powers of $v/v_\phi = kv/\omega_r \ll 1$.

### 6.2.2 Step-by-Step Derivation

**Step 1: Taylor-expand the resonant denominator.**

Write:

$$\frac{1}{v - \omega_r/k} = \frac{-k/\omega_r}{1 - kv/\omega_r}.$$

Since $kv/\omega_r \ll 1$ for most particles (those near the thermal core), expand as a geometric series:

$$\frac{1}{v - \omega_r/k} = -\frac{k}{\omega_r}\left[1 + \frac{kv}{\omega_r} + \left(\frac{kv}{\omega_r}\right)^2 + \left(\frac{kv}{\omega_r}\right)^3 + \cdots\right]. \tag{6.6}$$

This is a Taylor expansion of $1/(1-x)$ around $x = 0$ with $x = kv/\omega_r$. It is valid term-by-term for $|v| < \omega_r/k$, and the full series converges in the principal-value sense when integrated against $\partial \hat{f}_0/\partial v$ over all $v$.

**Step 2: Substitute into the principal value integral.**

$$\mathcal{P}\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0/\partial v}{v - \omega_r/k}\,dv = -\frac{k}{\omega_r}\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\left[1 + \frac{kv}{\omega_r} + \frac{k^2 v^2}{\omega_r^2} + \frac{k^3 v^3}{\omega_r^3} + \cdots\right]dv. \tag{6.7}$$

We now evaluate each term using **Integration by Parts (IBP)**: since $\partial \hat{f}_0/\partial v$ appears in the integrand, we shift the derivative onto the powers of $v$ using:

$$\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\cdot v^n\,dv = \left[\hat{f}_0\,v^n\right]_{-\infty}^{\infty} - n\int_{-\infty}^{\infty} \hat{f}_0\,v^{n-1}\,dv = -n\int_{-\infty}^{\infty} \hat{f}_0\,v^{n-1}\,dv,$$

where the boundary term vanishes because $\hat{f}_0 \to 0$ faster than any power of $v$ for $|v| \to \infty$ (Maxwellian decay). Each integral on the right is a **moment** of the distribution function:

- $n = 0$: $\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\,dv = [\hat{f}_0]_{-\infty}^{\infty} = 0.$ (The distribution vanishes at $\pm\infty$.)
- $n = 1$: $\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\cdot v\,dv = -\int_{-\infty}^{\infty} \hat{f}_0\,dv = -1.$ (Normalisation: $\int \hat{f}_0\,dv = 1$.)
- $n = 2$: $\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\cdot v^2\,dv = -2\int_{-\infty}^{\infty} \hat{f}_0\,v\,dv = 0.$ (The first moment $\int \hat{f}_0\,v\,dv = \bar{v} = 0$ for a zero-drift Maxwellian.)
- $n = 3$: $\int_{-\infty}^{\infty} \frac{\partial \hat{f}_0}{\partial v}\cdot v^3\,dv = -3\int_{-\infty}^{\infty} \hat{f}_0\,v^2\,dv = -3v_{Te}^2.$ (Second moment: $\int \hat{f}_0 v^2\,dv = v_{Te}^2 = T_e/m_e$.)

**Step 3: Assemble the expansion.**

Substituting these moment values into (6.7):

$$\mathcal{P}\int \frac{\partial \hat{f}_0/\partial v}{v - \omega_r/k}\,dv = -\frac{k}{\omega_r}\left[0 + \frac{k}{\omega_r}(-1) + \frac{k^2}{\omega_r^2}(0) + \frac{k^3}{\omega_r^3}(-3v_{Te}^2) + \cdots\right]$$

$$= -\frac{k}{\omega_r}\left[-\frac{k}{\omega_r} - \frac{3k^3 v_{Te}^2}{\omega_r^3} + \cdots\right] = \frac{k^2}{\omega_r^2} + \frac{3k^4 v_{Te}^2}{\omega_r^4} + \cdots \tag{6.8}$$

$$= \frac{k^2}{\omega_r^2}\left[1 + \frac{3k^2 v_{Te}^2}{\omega_r^2} + \cdots\right].$$

**Step 4: Substitute into $\varepsilon_R = 0$.**

$$\varepsilon_R = 1 - \frac{\omega_{pe}^2}{k^2}\cdot\frac{k^2}{\omega_r^2}\left[1 + \frac{3k^2 v_{Te}^2}{\omega_r^2}\right] = 0$$

$$\Longrightarrow\quad 1 - \frac{\omega_{pe}^2}{\omega_r^2}\left[1 + \frac{3k^2 v_{Te}^2}{\omega_r^2}\right] = 0$$

$$\Longrightarrow\quad \omega_r^2 = \omega_{pe}^2\left[1 + \frac{3k^2 v_{Te}^2}{\omega_r^2}\right]. \tag{6.9}$$

**Step 5: Iterate to lowest order in $k\lambda_D$.**

To leading order (zeroth order in $k\lambda_D$), equation (6.9) gives $\omega_r^2 \approx \omega_{pe}^2$, the cold-plasma result. Substituting this back into the correction term on the right (which is already a small correction of order $k^2\lambda_D^2$):

$$\omega_r^2 \approx \omega_{pe}^2\left[1 + \frac{3k^2 v_{Te}^2}{\omega_{pe}^2}\right] = \omega_{pe}^2 + 3k^2 v_{Te}^2.$$

Since $\lambda_D = v_{Te}/\omega_{pe}$, we may write $v_{Te}^2 = \omega_{pe}^2\lambda_D^2$, giving the standard form:

$$\boxed{\omega_r^2 = \omega_{pe}^2 + 3k^2 v_{Te}^2 = \omega_{pe}^2\!\left(1 + 3k^2\lambda_D^2\right).} \tag{6.10}$$

This is the **Bohm–Gross Dispersion Relation**.

### 6.2.3 Variable Anatomy and Physical Interpretation

| Symbol | Meaning | Units |
|--------|---------|-------|
| $\omega_r$ | Real oscillation frequency of the Langmuir wave | $\text{rad s}^{-1}$ |
| $\omega_{pe} = \sqrt{n_0 e^2/m_e\epsilon_0}$ | Electron plasma frequency — the fundamental resonant frequency of the electron gas against the ion background | $\text{rad s}^{-1}$ |
| $k$ | Wavenumber ($= 2\pi/\lambda$) — specifies the spatial scale of the perturbation | $\text{m}^{-1}$ |
| $v_{Te} = \sqrt{T_e/m_e}$ | Electron thermal velocity — the characteristic speed set by temperature | $\text{m s}^{-1}$ |
| $\lambda_D = v_{Te}/\omega_{pe}$ | Debye length — the natural length scale of the plasma | $\text{m}$ |

**Physical interpretation term by term:**

- $\omega_{pe}^2$: This is the **cold-plasma term**. It represents the restoring force of the electrostatic field pulling displaced electrons back toward the ion background. Even in a plasma at absolute zero, electrons would oscillate at $\omega_{pe}$. The plasma frequency depends only on density — denser plasma oscillates faster.

- $3k^2 v_{Te}^2 = 3k^2\omega_{pe}^2\lambda_D^2$: This is the **thermal correction** (the Bohm–Gross term). It represents the additional stiffness of the electron pressure — the tendency of a compressed, finite-temperature gas to push back. Shorter wavelengths (larger $k$) probe sharper compressions, where pressure gradients are steeper and the restoring force is stronger. The coefficient 3 is not arbitrary: it is the ratio of the adiabatic compressibility for a 1D electron gas ($\gamma_\text{ad} = 3$ for one degree of freedom), consistent with the fact that the electrons are compressed only in the direction of wave propagation.

This is essential becuase the Bohm–Gross relation shows that Langmuir waves are **dispersive**: their phase velocity $\omega_r/k = \omega_{pe}/k\,\sqrt{1 + 3k^2\lambda_D^2}$ increases with $k$. Hence, short-wavelength waves travel faster than long-wavelength ones.

### 6.2.4 Transition: From Frequency to Lifetime

The Bohm–Gross relation answers the question "at what frequency does the plasma oscillate?" but says nothing about how long the oscillation lasts. We assumed $\omega$ was real and computed its real part. But our Landau Prescription showed that $\omega$ must in general be complex, with an imaginary part $\gamma$ encoding growth or decay.

To find $\gamma$, we must use the imaginary part of the dielectric condition: equation (6.5). This is the subject of the next subsection.

---

## 6.3 The Landau Damping Rate

### 6.3.1 Background: Why Waves Die Without Collisions

Classical fluid theory predicts that a wave in a non-viscous, non-resistive medium can propagate forever without loss. The Bohm–Gross relation, derived from $\text{Re}[\varepsilon] = 0$ alone, appears to support this: it gives a real frequency, implying undamped oscillation.

Yet simulations — and laboratory experiments — show that Langmuir waves decay. This decay is *collisionless*: it occurs even in a plasma so dilute that particles essentially never bump into each other. The mechanism is purely kinetic — a wave-particle interaction mediated by the resonant denominator we studied in Section 6.1.

The key physical picture is as follows. In a Maxwellian distribution, the slope $\partial f_0/\partial v$ is negative everywhere: there are always more slow particles than fast ones. A particle slightly slower than the wave phase velocity $v_\phi$ is overtaken by the wave crest, which accelerates it forward — the particle gains energy from the wave. A particle slightly faster than $v_\phi$ outruns the wave crest, which decelerates it — the particle loses energy to the wave. Since there are more slow particles than fast ones near $v_\phi$ in a Maxwellian, the net energy flow is **from the wave to the particles**: the wave decays.

This is Landau Damping — a transfer of organised wave energy into the random thermal motion of resonant particles, achieved entirely through the coherent wave-particle interaction.

### 6.3.2 The "Small $\gamma$" Approach (Linearisation in the Damping Rate)

To extract $\gamma$ quantitatively, we exploit the hierarchy of scales established by the Bohm–Gross analysis: the wave oscillates many times before it decays, so $|\gamma| \ll \omega_r$. This means we can treat $\gamma$ as a small perturbation to $\omega_r$.

**Step 1: Write $\omega = \omega_r + i\gamma$.**

The full complex frequency $\omega = \omega_r + i\gamma$ satisfies $\varepsilon(k, \omega) = 0$. We write the dielectric function using the separation (6.3):

$$\varepsilon(k, \omega_r + i\gamma) = \varepsilon_R(k, \omega_r + i\gamma) + i\,\varepsilon_I(k, \omega_r + i\gamma) = 0. \tag{6.11}$$

**Step 2: Taylor-expand $\varepsilon$ around $\omega_r$.**

Since $\gamma \ll \omega_r$, expand to first order in $\gamma$:

$$\varepsilon(k, \omega_r + i\gamma) \approx \varepsilon(k, \omega_r) + i\gamma\left.\frac{\partial \varepsilon}{\partial \omega}\right|_{\omega_r} = 0. \tag{6.12}$$

Now note that at $\omega = \omega_r$ (the real frequency satisfying the Bohm–Gross relation), the real part of the dielectric function vanishes by construction: $\varepsilon_R(k, \omega_r) = 0$. Therefore:

$$\varepsilon(k, \omega_r) = i\,\varepsilon_I(k, \omega_r). \tag{6.13}$$

Similarly, the derivative $\partial\varepsilon/\partial\omega$ at the real frequency $\omega_r$ is dominated by its real part (since $\varepsilon_I$ is exponentially small for a Maxwellian in the long-wavelength limit):

$$\left.\frac{\partial \varepsilon}{\partial \omega}\right|_{\omega_r} \approx \left.\frac{\partial \varepsilon_R}{\partial \omega}\right|_{\omega_r}. \tag{6.14}$$

**Step 3: Solve for $\gamma$.**

Substituting (6.13) and (6.14) into (6.12):

$$i\,\varepsilon_I + i\gamma\,\frac{\partial \varepsilon_R}{\partial \omega} = 0 \quad\Longrightarrow\quad \boxed{\gamma = -\frac{\varepsilon_I}{\partial\varepsilon_R/\partial\omega}\Bigg|_{\omega = \omega_r}.} \tag{6.15}$$

This is the central formula for the Landau damping rate. It expresses $\gamma$ as a ratio: the imaginary part of the dielectric function (which captures the resonant particle contribution) divided by the sensitivity of the real part to frequency (which quantifies the wave's stored energy).

We now evaluate the numerator and denominator separately.

### 6.3.3 Part A — The Numerator: Evaluating $\varepsilon_I$

From equation (6.5):

$$\varepsilon_I = -\frac{\pi\,\omega_{pe}^2}{k^2}\left.\frac{\partial \hat{f}_0}{\partial v}\right|_{v = \omega_r/k}. \tag{6.16}$$

We need $\partial \hat{f}_0/\partial v$ at the resonant velocity $v = \omega_r/k$. The equilibrium distribution is a **Maxwellian**:

$$f_0(v) = \frac{n_0}{\sqrt{2\pi}\,v_{Te}}\exp\!\left(-\frac{v^2}{2v_{Te}^2}\right), \qquad \hat{f}_0(v) = \frac{f_0}{n_0} = \frac{1}{\sqrt{2\pi}\,v_{Te}}\exp\!\left(-\frac{v^2}{2v_{Te}^2}\right). \tag{6.17}$$

Why the Maxwellian? It is the **maximum entropy** distribution: the unique distribution that maximises entropy (randomness) subject to fixed total particle number and total kinetic energy. Any plasma in thermal equilibrium — where particles have had time to exchange energy through collisions — settles onto a Maxwellian. It is characterised entirely by the density $n_0$ and the temperature $T_e$ (through $v_{Te}^2 = T_e/m_e$).

**Differentiate** $\hat{f}_0$ with respect to $v$ using the chain rule:

$$\frac{\partial \hat{f}_0}{\partial v} = \frac{1}{\sqrt{2\pi}\,v_{Te}}\cdot\exp\!\left(-\frac{v^2}{2v_{Te}^2}\right)\cdot\left(-\frac{v}{v_{Te}^2}\right) = -\frac{v}{v_{Te}^2}\,\hat{f}_0(v). \tag{6.18}$$

This result — that the slope of the Maxwellian is $-v/v_{Te}^2$ times the Maxwellian itself — is physically meaningful: the slope is negative for $v > 0$ (the distribution falls off for higher speeds), and zero at $v = 0$ (the peak).

**Evaluate at the resonant velocity** $v = \omega_r/k$:

$$\left.\frac{\partial \hat{f}_0}{\partial v}\right|_{v=\omega_r/k} = -\frac{\omega_r/k}{v_{Te}^2}\cdot\hat{f}_0\!\left(\frac{\omega_r}{k}\right) = -\frac{\omega_r}{k\,v_{Te}^2}\cdot\frac{1}{\sqrt{2\pi}\,v_{Te}}\exp\!\left(-\frac{\omega_r^2}{2k^2 v_{Te}^2}\right). \tag{6.19}$$

**Substitute into (6.16):**

$$\varepsilon_I = -\frac{\pi\,\omega_{pe}^2}{k^2}\cdot\left(-\frac{\omega_r}{k\,v_{Te}^2}\cdot\frac{1}{\sqrt{2\pi}\,v_{Te}}\right)\exp\!\left(-\frac{\omega_r^2}{2k^2v_{Te}^2}\right)$$

$$= \frac{\pi\,\omega_{pe}^2\,\omega_r}{k^3\,v_{Te}^2\,\sqrt{2\pi}\,v_{Te}}\exp\!\left(-\frac{\omega_r^2}{2k^2v_{Te}^2}\right) = \sqrt{\frac{\pi}{2}}\cdot\frac{\omega_{pe}^2\,\omega_r}{k^3\,v_{Te}^3}\exp\!\left(-\frac{\omega_r^2}{2k^2v_{Te}^2}\right). \tag{6.20}$$

### 6.3.4 Part B — The Denominator: Evaluating $\partial\varepsilon_R/\partial\omega$

We need the derivative of the real dielectric function with respect to $\omega$, evaluated at $\omega_r$. We use the **cold-plasma (leading-order) approximation** for $\varepsilon_R$:

$$\varepsilon_R \approx 1 - \frac{\omega_{pe}^2}{\omega^2}. \tag{6.21}$$

**Why is this valid here?** The correction from thermal effects in $\varepsilon_R$ is of order $k^2\lambda_D^2$ (as shown by the Bohm–Gross derivation). Since we are computing a correction $\gamma$ that is already exponentially small (much smaller than $k^2\lambda_D^2$ in the long-wavelength regime), using the leading-order $\varepsilon_R$ introduces only negligible errors.

Differentiating (6.21):

$$\frac{\partial\varepsilon_R}{\partial\omega} = \frac{2\omega_{pe}^2}{\omega^3}. \tag{6.22}$$

**Evaluate at $\omega_r \approx \omega_{pe}$** (to leading order in $k\lambda_D$):

$$\left.\frac{\partial\varepsilon_R}{\partial\omega}\right|_{\omega_r} \approx \frac{2\omega_{pe}^2}{\omega_{pe}^3} = \frac{2}{\omega_{pe}}. \tag{6.23}$$

**Physical meaning of this denominator.** The quantity $\omega_{pe}/2 \cdot \partial\varepsilon_R/\partial\omega$ is related to the **electromagnetic energy stored in the wave**. This is not a coincidence: for electrostatic waves, the wave energy density is proportional to $\partial(\omega\varepsilon_R)/\partial\omega \cdot |E|^2/2$. A steeper $\partial\varepsilon_R/\partial\omega$ means more energy is stored per unit field amplitude, so the wave is harder to damp — i.e., $\gamma$ is smaller for the same $\varepsilon_I$. The ratio $\varepsilon_I/(\partial\varepsilon_R/\partial\omega)$ is precisely the balance between "how fast energy is being extracted" (numerator) and "how much energy is stored" (denominator).

### 6.3.5 Step 4: The Exponent Correction from Bohm–Gross

Before combining the numerator and denominator, we refine the exponent in $\varepsilon_I$ using the full Bohm–Gross frequency (6.10). The argument of the exponential is:

$$\frac{\omega_r^2}{2k^2 v_{Te}^2} = \frac{\omega_{pe}^2 + 3k^2 v_{Te}^2}{2k^2 v_{Te}^2} = \frac{\omega_{pe}^2}{2k^2 v_{Te}^2} + \frac{3}{2} = \frac{1}{2k^2\lambda_D^2} + \frac{3}{2}. \tag{6.24}$$

where we used $\lambda_D^2 = v_{Te}^2/\omega_{pe}^2$. The extra $3/2$ in the exponent comes from the thermal correction to the wave frequency — it makes the phase velocity slightly higher than $\omega_{pe}/k$, pushing the resonance deeper into the tail of the Maxwellian, and making the damping rate exponentially *smaller*.

Therefore:

$$\exp\!\left(-\frac{\omega_r^2}{2k^2 v_{Te}^2}\right) = \exp\!\left(-\frac{1}{2k^2\lambda_D^2} - \frac{3}{2}\right). \tag{6.25}$$

### 6.3.6 Step 5: Assembling the Landau Damping Rate

Substituting (6.20), (6.23), and (6.25) into (6.15):

$$\gamma = -\frac{\varepsilon_I}{\partial\varepsilon_R/\partial\omega} = -\frac{\sqrt{\pi/2}\cdot\omega_{pe}^2\,\omega_r/(k^3 v_{Te}^3)\cdot\exp(-\omega_r^2/2k^2v_{Te}^2)}{2/\omega_{pe}}.$$

Using $\omega_r \approx \omega_{pe}$ and $v_{Te}^3 = \omega_{pe}^3\lambda_D^3$:

$$\gamma = -\sqrt{\frac{\pi}{2}}\cdot\frac{\omega_{pe}^2\cdot\omega_{pe}}{k^3\,\omega_{pe}^3\,\lambda_D^3}\cdot\frac{\omega_{pe}}{2}\cdot\exp\!\left(-\frac{1}{2k^2\lambda_D^2} - \frac{3}{2}\right)$$

$$= -\sqrt{\frac{\pi}{2}}\cdot\frac{1}{2}\cdot\frac{\omega_{pe}}{k^3\lambda_D^3}\cdot\exp\!\left(-\frac{1}{2k^2\lambda_D^2} - \frac{3}{2}\right)$$

$$\boxed{\gamma = -\sqrt{\frac{\pi}{8}}\,\cdot\,\frac{\omega_{pe}}{k^3\lambda_D^3}\,\cdot\,\exp\!\left(-\frac{1}{2k^2\lambda_D^2} - \frac{3}{2}\right).} \tag{6.26}$$

### 6.3.7 Variable Anatomy and Physical Interpretation

| Symbol | Meaning | Effect on $\gamma$ |
|--------|---------|-------------------|
| $\gamma < 0$ | Negative imaginary frequency → **wave decays** in time as $e^{\gamma t}$ | — |
| $\omega_{pe}$ | Plasma frequency — sets the fundamental timescale | $|\gamma| \propto \omega_{pe}$: denser plasma oscillates and damps faster |
| $k^3\lambda_D^3$ | Dimensionless wavenumber cubed | $|\gamma| \propto 1/k^3\lambda_D^3$: long-wavelength waves ($k\lambda_D \ll 1$) damp very slowly |
| $\exp(-1/2k^2\lambda_D^2)$ | Exponential suppression | **Dominant factor**: damping is exponentially weak at long wavelengths |
| $\exp(-3/2)$ | Bohm–Gross correction | Moves resonance deeper into the Maxwellian tail, reducing $|\gamma|$ by a fixed factor |

**The exponential dependence is the most important feature of (6.26).** As $k\lambda_D \to 0$ (very long waves), $\exp(-1/2k^2\lambda_D^2) \to 0$ faster than any power of $k$. This means long-wavelength Langmuir waves are essentially undamped — they propagate almost indefinitely. Short-wavelength waves (large $k\lambda_D$), on the other hand, damp on timescales comparable to an oscillation period.

The physical reason is that the resonant velocity $v_\phi = \omega_r/k \gg v_{Te}$ for small $k\lambda_D$: the wave is so fast that essentially no thermal particles can keep up with it. The Maxwellian distribution has an exponentially small number of particles at $v = v_\phi$, so the damping is exponentially weak. This is why the damping rate is not a power law in $k\lambda_D$ but an exponential — the resonant particle density is exponentially small.

### 6.3.8 Why This Result is Non-perturbative

It is worth pausing to appreciate how remarkable equation (6.26) is. The Landau damping rate contains the factor $\exp(-1/2k^2\lambda_D^2)$, which **cannot be obtained by any order of perturbation theory** in the small parameter $k\lambda_D$. No matter how many terms of a Taylor series in $k\lambda_D$ you compute, you will never reproduce an exponential. This is why Landau damping was overlooked for so long, and why Landau's initial-value approach was so revolutionary: it is an intrinsically non-perturbative result, arising from the analytic structure of the dielectric function in the complex frequency plane.

## Summary

The three results of Section 6 form a single, integrated argument:

| Subsection | Result | Physical Role |
|-----------|--------|---------------|
| 6.1 — Landau Prescription | Plemelj–Sokhotski: $\int \to \mathcal{P}\int + i\pi g(v_\phi)$ | Correctly separates the resonant (energy-exchanging) and bulk (frequency-setting) particle contributions |
| 6.2 — Bohm–Gross Relation | $\omega_r^2 = \omega_{pe}^2 + 3k^2 v_{Te}^2$ | Determines the real oscillation frequency — the "pitch" of the Langmuir wave |
| 6.3 — Landau Damping Rate | $\gamma = -\sqrt{\pi/8}\,(\omega_{pe}/k^3\lambda_D^3)\,e^{-1/2k^2\lambda_D^2 - 3/2}$ | Determines the wave lifetime — how fast the "song" fades |

**The logical thread is:**

Section 5 gave us the condition $\varepsilon(k,\omega) = 0$, but could not evaluate the singular integral it contained.

$\longrightarrow$ Section 6.1 (Landau Prescription) resolved the singularity, revealing that $\varepsilon$ splits into a real part (from the bulk response) and an imaginary part (from the resonant particles).

$\longrightarrow$ Section 6.2 (Bohm–Gross) solved $\text{Re}[\varepsilon] = 0$ using a long-wavelength Taylor expansion to find $\omega_r(k)$.

$\longrightarrow$ Section 6.3 (Landau Damping) used $\text{Im}[\varepsilon]$ and the derivative of $\varepsilon_R$ to find $\gamma(k)$, completing the full complex dispersion relation $\omega(k) = \omega_r(k) + i\gamma(k)$.

Together, they provide the complete theoretical prediction for the Langmuir wave — a prediction that can be directly verified against the Vlasov–Poisson simulation that follows.
